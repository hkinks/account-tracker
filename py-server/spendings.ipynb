```python
from pathlib import Path
import os

folder_dir = Path(r"C:\Users\hkink\OneDrive\Personal\Finances\Konto väljavõtted")
files = folder_dir.glob("*palk.csv")
# file_path = next(files)
# print(file_path)
```

```python
import pandas as pd

# file_path = 'bank_statement.csv'
all_df = None
for file_path in files:
    df = pd.read_csv(file_path, sep=';', encoding='utf-8')
    if all_df is None:
        all_df = df
    df = df[["Kuupäev", "Saaja/Maksja", "Selgitus", "Summa", "Valuuta", "Kliendi konto"]].copy()
    df = df[~df['Selgitus'].isin(['Käive'])].copy()
    all_df = pd.concat([all_df, df])

df = all_df
df
```

```python
# Convert date column to datetime
df['Kuupäev'] = pd.to_datetime(df['Kuupäev'], errors='coerce', format='%d.%m.%Y')
# df.dtypes

# # Filter for debit (spending) transactions
# df = df[df['Deebet/Kreedit'] == 'Deebet']

# Extract year and month from date
df['Month'] = df['Kuupäev'].dt.to_period('M')

# # Convert "Summa" to numeric and handle possible conversion issues
df['Summa'] = df['Summa'].str.replace(',', '.')
df['Summa'] = pd.to_numeric(df['Summa'], errors='coerce')

df.sort_values(by='Kuupäev', ascending=False)
```

```python
# Remap columns
df = df.rename(columns={"Kuupäev": "Date", "Saaja/Maksja": "Sender/Receiver", "Selgitus": "Description", "Summa": "Amount", "Valuuta": "Currency", "Kliendi konto": "Account"})
```

```python
import numpy as np
df["Receiver"] = np.where(df["Amount"]>0, df["Account"], df["Sender/Receiver"])
df["Sender"] = np.where(df["Amount"]<0, df["Account"], df["Sender/Receiver"])

df
```

```python
import os

from sqlalchemy import create_engine, Column, Integer, String, Numeric
from sqlalchemy.orm import sessionmaker
from dotenv import load_dotenv
load_dotenv(override=True)

# Database connection
db_url = os.getenv("DB_URL")
engine = create_engine(db_url)
SessionLocal = sessionmaker(engine)

# Insert to postgres
from model import BankTransaction

for _, row in df.iterrows():
    session = SessionLocal()
    transaction = BankTransaction(description=row["Description"], 
                                  amount=row["Amount"],
                                  sender=row["Sender"],
                                  receiver=row["Receiver"],
                                  date=row["Date"],
                                  currency=row["Currency"]

                                  )
    session.add(transaction)
    try:
        session.commit()
    except:
        continue
    session.close()
```

```python
# # Group by month and calculate sum of spending
monthly_spending = df.groupby('Month')['Summa'].sum()

montly_spending
```

```python
# Find the month with the biggest spending
max_spending_month = monthly_spending.idxmax()
max_spending_value = monthly_spending.max()

print(f'Biggest spending month: {max_spending_month}, Amount: {max_spending_value}')

# Display aggregated monthly spendings
print(monthly_spending)
```
